```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the LCOM CSV output file
lcom_df = pd.read_csv('lcom_output/TypeMetrics.csv')  # adjust path as needed
lcom_df
```

```python
# Optionally, create a grouped bar chart if you have multiple metrics (e.g., LCOM1, LCOM2, etc.)
# Assuming the CSV has columns: ClassName, LCOM1, LCOM2, LCOM3, LCOM4, LCOM5, YALCOM
lcom_df_new=lcom_df[80:100]
metrics = ['LCOM1', 'LCOM2', 'LCOM3', 'LCOM4', 'LCOM5', 'YALCOM']
lcom_df_new = lcom_df_new.set_index('Type Name')

# Plot a grouped bar chart
lcom_df_new[metrics].plot(kind='bar', figsize=(12, 8))
plt.xlabel('Class Name')
plt.ylabel('LCOM Value')
plt.title('LCOM Values for Selected Java Classes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
